```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
sc = [1.0, 0.965, 0.835, 0.655, 0.505, 0.290, 0.124, 0.038, 0.006, 0.001]
ss = [1.0, 0.981, 0.882, 0.701, 0.514, 0.327, 0.135, 0.025, 0.013, 0.001]
x1 = [0.7, 0.6, 0.5, 0.4, 0.3,  0.25, 0.2, 0.15, 0.1, 0.05]
diff = np.array(ss) - np.array(sc)
x1.reverse()
```

```python
plt.plot(x1, sc, label = 'sc')
plt.plot(x1, ss, label = 'ss')
plt.plot(x1,diff)
plt.fill_between(x1, diff, color='blue')
plt.xlabel('% of blocks')
plt.ylabel('% pig wins')
plt.legend()
```
